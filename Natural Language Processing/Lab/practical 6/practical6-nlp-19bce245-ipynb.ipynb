```python
b_topAuthors = list(a1.loc[a1['publication'] == 'Breitbart']['author'].value_counts()[: 10].index)
```

```python
c_topAuthors = list(a1.loc[a1['publication'] == 'CNN']['author'].value_counts()[:10].index)
```

## Data Preprocessing

```python
data = pd.DataFrame()
```

```python
for i in b_topAuthors:
    data = pd.concat([data, a1.loc[a1['author'] == i][:100]])
for j in c_topAuthors:
    data = pd.concat([data, a1.loc[a1['author'] == j][:100]])
```

```python
data['author'].value_counts().plot(kind='barh')
```

```python
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=3000, stop_words='english') 
input = vectorizer.fit_transform(data['content'])
```

```python
input
```

```python
input.toarray()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
output = le.fit_transform(data['author'])
```

```python
output
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input, output, test_size=0.3, random_state=42)
```

```python
from sklearn import datasets, metrics, neighbors
from sklearn.model_selection import train_test_split, GridSearchCV 
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
hyperparams = {'n_neighbors' : [1, 5, 10, 25, 50, 100], 'weights' : ['uniform', 'distance']} 
clf = neighbors.KNeighborsClassifier()
clf_gs = GridSearchCV(clf, hyperparams, cv=5) 
clf_gs.fit(X_train, y_train)
print(clf_gs.best_params_)
```

```python
prediction = clf_gs.predict(X_test)
# printing accuracy of the model
print("Accuracy: ",metrics.accuracy_score(y_test, prediction, normalize=True)) # printing classification report of the model
# print(metrics.classification_report(y_test,prediction))
# printing confusion matrix of the model
 
```
