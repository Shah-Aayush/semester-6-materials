```python
import numpy as np
import pandas as pd

train_data = pd.read_csv('train.csv')
train_data.head(10)
```

```python
test_data = pd.read_csv('test.csv')
test_data.head(10)
```

```python
train_data.isnull().sum()
```

```python
train_data['Age'].fillna(train_data['Age'].mean() , inplace = True)
train_data.head(10)
```

```python
print("test_data : " , test_data.shape)
```

```python
train_data.drop(['PassengerId' , 'Name' , 'Ticket' , 'Fare' , 'Cabin'] , axis = 1 , inplace = True)
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]
X = pd.get_dummies(train_data[["Pclass", "Sex", "SibSp", "Parch"]])
X_test = pd.get_dummies(test_data[["Pclass", "Sex", "SibSp", "Parch"]])

model = RandomForestClassifier()
model.fit(X, y)
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({"PassengerId": test_data.PassengerId, "Survived": predictions})
```

```python
print(output)
output.to_csv('19BCE245.csv', index = False)
```
