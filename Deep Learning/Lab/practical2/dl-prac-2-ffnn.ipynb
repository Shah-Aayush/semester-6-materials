```python
import pandas as pd
import numpy as np 
import tensorflow as tf 

from tensorflow import keras #we are going to code using keras
from keras.models import Sequential, Model 
from keras.layers import Dense, BatchNormalization, Activation, Input #Importing layers which we will use in the model
from keras.utils.np_utils import to_categorical #for encoding target labels
from keras.callbacks import ReduceLROnPlateau
from keras.utils import np_utils
```

```python
train = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
test = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
submission = pd.read_csv("/kaggle/input/digit-recognizer/sample_submission.csv")
```

```python
train
```

```python
shape = (28, 28, 1)
image = lambda data: np.array(data.drop(columns=['label'], errors='ignore'))
label = lambda data: data['label']

x_train=np.array(train.drop(columns=['label'],errors='ignore'))
y_train=train['label']
print(y_train)
```

```python
x_train=x_train/255 #Normalizing values
y_train= np_utils.to_categorical(y_train) #It will convert the y_train into 42000 rows and 10 columns(because of total classes 0-9) and will assign 1 to the present value and 0 to absent value.
```

```python
print(y_train)
```

```python
model = Sequential()

model.add(Dense(units=512,activation='relu',input_shape=(784,)))
model.add(Dense(units=256,activation='relu'))
# model.add(Dense(units=128,activation='relu'))
# model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=10,activation='softmax'))

tf.keras.utils.plot_model(model)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
reduceLROnPlateau=ReduceLROnPlateau(monitor='val_acc', patience=3, verbose=1, factor=0.5, min_lr=0.00001)
model.fit(x=x_train, y=y_train, epochs=60, batch_size=120, validation_split=0.2)
```

```python
y_pre = model.predict(test)
y_pre = y_pre.argmax(axis = 1)
sub = pd.DataFrame({"ImageId":test.index + 1,"Label":y_pre})
sub.to_csv('19BCE226_FFNN_submission.csv',index=False)
```
