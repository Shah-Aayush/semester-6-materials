```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)import numpy as np # linear algebra
```

```python
train_df=pd.read_csv("../input/digit-recognizer/train.csv")
test_df=pd.read_csv("../input/digit-recognizer/test.csv")
```

```python
y_train=train_df.pop('label')
I_train=np.array(train_df)
I_test=np.array(test_df)
```

```python
X_train=I_train.reshape(-1,28,28,1)
X_test=I_test.reshape(-1,28,28,1)
```

```python
print(X_train.max())
X_train.shape
```

```python
X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255
```

```python
import keras
from keras.models import Model
from keras.layers import Dense, Input,Dropout
from keras.layers import Conv2D,MaxPooling2D,Activation,Flatten,BatchNormalization
from keras.utils import np_utils
from tensorflow.keras.models import Sequential, Model
```

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28,1)))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x=X_train,y=y_train,batch_size=120,epochs=50)
model.summary()
```

```python
y_pred = model.predict(X_test)
y_pred = y_pred.argmax(axis=-1)
y_pred
```

```python
#SUBMITTING PREDICTIONS
sub=pd.read_csv('/kaggle/input/digit-recognizer/sample_submission.csv', header='infer')
sub["Label"]=y_pred
sub.to_csv('submission.csv', index=False)
```
